```python
################################
### Import Stuff
################################
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
################################
### Define functions
################################

#SI is both S and I. S is SI[0] and I is SI[1]

def SIS(SI, t, a_hi, a_lo, b):
    """Return the time-dependent growth rate of the susceptible and infected populations"""
    
    a = ((a_hi-a_lo)/2)*np.sin(2*np.pi*t/365) + (a_hi + a_lo)/2
    
    dSdt = -a*SI[0]*SI[1] + b*SI[1]
    dIdt = a*SI[0]*SI[1] - b*SI[1]
    return np.array([dSdt, dIdt])
```

```python
################################
### Set parameters
################################


a_hi = 0.002
a_lo = 0.0000
b = 0.15

max_time = 1000 #days
num_time_points = 10000



################################
### Initialze simulation
################################
N = 1000
I_o = 1
S_o = N - I_o

SI_o = np.array([S_o, I_o])

t = np.linspace(0,max_time,num_time_points)

################################
### Run the simulation
################################

SI = odeint(SIS, SI_o, t, args=(a_hi, a_lo, b))


################################
### Plot the results
################################

plt.figure()
plt.plot(t,SI[:,0], label='S')
plt.plot(t,SI[:,1], label='I')
plt.xlabel('time')
plt.ylabel('pop size')
plt.legend()
```
