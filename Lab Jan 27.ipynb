Code a 1D, unbiased, random walk in which the step length decreases by a factor of each time step. Make the number of walkers and the number of time steps variables.

```python
#imports 

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#variables

num_walkers = 100
num_time_steps = 50
p = 0.5

random_walk = np.zeros((num_walkers,num_time_steps))

for t in range(1,num_time_steps):#loop over time
    for w in range(num_walkers):#loop over walkers
        step_length = random.choice([1,-1])*(p**t)#choose step length
        random_walk[w,t] = random_walk[w,t-1] + step_length
        
#plot 
plt.figure()
_=plt.plot(np.transpose(random_walk[:,:]))
```
