```python
#Code Rule 30 of Wolfram's elementary CAs for a 1D lattice with periodic boundary conditions. 
#Store/Plot the result as a 2D array, with length on one axis and time on the other. 
#Start with random initial conditions. Does it look like a seashell?

# import 

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# let x be a 10 by 10 2d array
size = 100
x = np.empty([size, size])
x.fill(np.nan)
print(x)
```

```python
for i in range(size): #start the first row with random conditions
    x[0,i] = random.choice([0,1])
print(x)
```

```python
def update_rules(a):
    b = np.empty([size, 1])
    b.fill(np.nan)
    for i in range(size):
        if i==0:
            left = a[-1]
        if i==size-1:
            right = a[0]
        else:
            left = a[i-1]
            right = a[i+1]
        center = a[i]
        if (left == 0) & (center == 0) & (right == 0):
            b[i] = 0
        elif (left == 1) & (center == 0) & (right == 1):
            b[i] = 0
        elif (left == 1) & (center == 1) & (right == 0):
            b[i] = 0
        elif (left == 1) & (center == 1) & (right == 1):
            b[i] = 0
        else:
            b[i] = 1
    return b
            
```

```python
for i in range(size-1):
    x[i+1,:] = update_rules(x[i,:])[:,0]


print(x)
```
