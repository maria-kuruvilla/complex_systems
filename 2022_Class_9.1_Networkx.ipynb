```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx ############## package for creating and working with networks!
```

```python

### create graph
G = nx.Graph()

### add nodes
#G.add_node(0) #individually
G.add_nodes_from([1, 2, 3])

### add edges
G.add_edge(1, 2)
G.add_edge(3, 2)
G.add_edge(1, 3)
G.add_edge(4, 2)

### visualize
nx.draw(G, with_labels=True)
```

```python
####look at network properties
G.nodes()
G.edges()
G.degree()
```

```python
#### create graph from adacency matrix
N = 4
adj_mat = np.random.randint(0,2,(N,N))
G2 = nx.from_numpy_matrix(adj_mat)

print(adj_mat)
# visualize
nx.draw(G2, with_labels=True)
```

# built-in tools for creating cannonical graphs

```python
# Erdos Renyi (random)

N = 32
link_prob = .2
G_ER = nx.erdos_renyi_graph(N, link_prob)
nx.draw(G_ER, with_labels=False)
```

```python
# Watts Strogatz (Small world)

N = 32
k = 4
prob_link_shuffle = .1
G_WS = nx.watts_strogatz_graph(N, k, prob_link_shuffle)
nx.draw(G_WS, with_labels=False)

```

```python
# Barabasi Albert (scale free)


N = 1000
m = 1
G_BA = nx.barabasi_albert_graph(N, m)
nx.draw(G_BA, with_labels=False)


```

```python
# look at degree distribution
degree_list = [d for n, d in G_BA.degree()]
plt.figure()
bar_x,hist_y = np.unique(degree_list, return_counts=True)
mean_degree = np.mean(degree_list)
pdf_y = hist_y/hist_y.sum()
plt.bar(bar_x,pdf_y)
plt.figure()
plt.loglog(bar_x,pdf_y)
```
