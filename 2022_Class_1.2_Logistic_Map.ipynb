```python
################################
### Import Stuff
################################
import numpy as np
import matplotlib.pyplot as plt
```

```python
################################
### Set parameters
################################

max_time = 100
r = 4
x_0 = 0.65

################################
### Create recording variables
################################
x_over_time = np.zeros(max_time)

################################
### Initialze simulation
################################
x_over_time[0] = x_0
#print(x_over_time)


################################
### Run the simulation
################################

for t in range(1,max_time):
    #print(t)
    x_over_time[t] = r*(1-x_over_time[t-1])*x_over_time[t-1] #r*(1-x)*x
    
    
################################
### Plot the results
################################

plt.figure()
plt.plot(x_over_time)
plt.xlabel('time')
plt.ylabel('pop size')
```
