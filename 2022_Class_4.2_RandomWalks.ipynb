```python
import numpy as np
import matplotlib.pyplot as plt
```

# Single walker

```python
num_time_steps = 1000

x_over_time = np.zeros(num_time_steps)

for t in range(1,num_time_steps):
    step = np.random.choice([-1,1])
    #print(step)
    x_over_time[t] = x_over_time[t-1] + step
    
plt.figure()
plt.plot(x_over_time)
```

# Multiple walkers

```python
num_time_steps = 1000
num_walkers = 100

x_over_time = np.zeros((num_walkers,num_time_steps))

for t in range(1,num_time_steps):
    for walker in range(num_walkers):
        step = np.random.choice([-1,1])
        #print(step)
        x_over_time[walker,t] = x_over_time[walker,t-1] + step
    
plt.figure()
_ = plt.plot(np.transpose(x_over_time))
```

# Vectorize loop over walkers

```python
num_time_steps = 1000
num_walkers = 100

x_over_time = np.zeros((num_walkers,num_time_steps))

for t in range(1,num_time_steps):
    steps = np.random.choice([-1,1],size=num_walkers)
    #print(step)
    x_over_time[:,t] = x_over_time[:,t-1] + steps
    
plt.figure()
_ = plt.plot(np.transpose(x_over_time))
```

# Vectorize loop over time

```python
num_time_steps = 1000
num_walkers = 100

x_over_time = np.zeros((num_walkers,num_time_steps))

steps = np.random.choice([-1,1],size=(num_walkers,num_time_steps))

x_over_time = np.cumsum(steps,axis=1)


    
plt.figure()
_ = plt.plot(np.transpose(x_over_time))
```
