```python
#Use a for loop to simulate the discrete-time formulation. 
#Use the odeint() function from the scipi.integrate library to simulate the continuous time formulation. 
#Are they the same?

################################
### Import Stuff
################################
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#using for loop to simulate the discrete-time formulation

#Setting parameters

r = 3
max_time = 100
n_0 = 3
K=50

################################
### Create recording variables
################################
n_over_time = np.zeros(max_time)

################################
### Initialze simulation
################################
n_over_time[0] = n_0
#print(n_over_time)
```

```python
for t in range(1,max_time):
    #print(t)
    n_over_time[t] = r*(1-n_over_time[t-1]/K)*n_over_time[t-1] + n_over_time[t-1]#r*(1-x)*x
    #print(n_over_time[t])
################################
### Plot the results
################################

plt.figure()
plt.plot(n_over_time)
plt.xlabel('time')
plt.ylabel('pop size')
```

```python
#Use the odeint() function from the scipi.integrate library to simulate the continuous time formulation. 
#Are they the same?

def logistic(n, t, r, K):
    
    dndt = r*n*(1-n/K)
    return dndt
```

```python
num_time_points = 100
t = np.linspace(0,max_time,num_time_points)
n = odeint(logistic, n_0, t, args=(r, K))
```

```python
################################
### Plot the results
################################

plt.figure()
plt.plot(t,n)

plt.xlabel('time')
plt.ylabel('pop size')
plt.legend()
```
