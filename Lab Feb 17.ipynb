```python
# code  Bak & Sneppen model

#imports 

import numpy as np
import matplotlib.pyplot as plt

```

```python
# set up

num_species = 50
max_time_steps = 1000


#give random numbers to each species
fitness_landscape = np.random.random(num_species)
#print(fitness_landscape)

#array to keep track of min fitness
min_fitness = np.zeros(max_time_steps)

final_array = fitness_landscape
for i in range(max_time_steps):#loop over time
    
    
    a = np.where(fitness_landscape==np.min(fitness_landscape))[0][0]#find the index of species with minimum fitness
    #print(a)
    #keeping track of min fitness
    min_fitness[i] = fitness_landscape[a]
    fitness_landscape[a] = np.random.random(1)#pick random number for that species

    fitness_landscape[a-1] = np.random.random(1)
    fitness_landscape[(a+1)%num_species] = np.random.random(1)#pick random number for neighbors
    final_array = np.column_stack((final_array,fitness_landscape))
    
    # if adequate time has passed
    #histogram of fitness
    n_bins = 10
    plt.hist(fitness_landscape, bins = n_bins)
    
final_array_t = np.transpose(final_array)
# print(final_array)
# print(np.log(final_array+1))
plt.figure()
plt.imshow(np.log(final_array_t+1), extent=[0,num_species,max_time_steps,0], aspect=num_species/max_time_steps)
```
