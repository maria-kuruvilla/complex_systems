```python
#imports

import numpy as np
import matplotlib.pyplot as plt

```

```python
L = 20
x = np.arange(-L,L)
gamma = [0.5,1,2]
plt.figure()
for g in gamma:
    
    p = np.abs(x/L)**g
    plt.plot(x,p, label = str(g))
plt.legend(title = 'gamma')
```

```python
#probability to move left

L = 50
x = np.arange(-L,L)
gamma = [0.25,0.5,1,2,3,4]
plt.figure()
for g in gamma:
    
    p = np.abs(x/L)**g
    p_left = (np.sign(x)*np.abs(x/L)**g + 1)/2
    plt.plot(x,p_left, label = str(g))
plt.legend(title = 'gamma')
```

```python
#code simulation with p_left

########################
### parameters
########################

num_time_steps = 10
L = 50
x = np.arange(-L,L)
gamma = 1
p_left = (np.sign(x)*np.abs(x/L)**gamma + 1)/2
alpha = 0.5


N = np.zeros(2*L) #make our list of site densities
N[0] = 1 #put all of our walker density in a middle site
print(N)
print(x)
plt.figure()
plt.bar(x,N)
```

```python
########################
### run
########################
for t in range(num_time_steps):
    stay = (1-alpha)*N
    left = alpha*N/2
    right = alpha*N/2
    
    N = stay + np.roll(left,-1) + np.roll(right,1)
```
